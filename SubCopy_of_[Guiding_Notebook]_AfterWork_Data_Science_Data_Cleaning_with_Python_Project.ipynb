```python
# Lets first start by creating a copy of our dataframe
# df_clean = df.copy(). We will use this copy as our cleaning copy.
# ---
#
df_clean = data_no_duplicates.copy()
df_clean.head()
```

```python
# Then extracting the lowest delay time in the column how_long_delayed from the string
#

df_clean['How_Long_Delayed'] = df_clean['How_Long_Delayed'].str.extract('(\d+)')
df_clean['How_Long_Delayed'].head(281110)

#pd.reset_option('display.max_rows')

```

We impute the null values in 'how_long_delayed' column with mean of the column. This will take a couple of steps...

```python
# We first convert our how_long_delayed to float type to allow for imputation
# ---
#

df_clean['How_Long_Delayed'] = df_clean['How_Long_Delayed'].str.extract('(\d+)').astype(float)
```

```python
# Then later perform our mean imputation
# ---
#
print('Check the type of the column How_Long_Delayed: ', df_clean['How_Long_Delayed'].dtype)
print(df_clean['How_Long_Delayed'])
# I would like to replace only the 0 MIN value by the MIN
mean_delay=df_clean['How_Long_Delayed'].mean()

print('The Mean is:', mean_delay)

#or
# one way maybe to proceed  is to  compute the mean without considering the 0 MIN value

#mean_delay = df_clean[df_clean['How_Long_Delayed'] != 0]['How_Long_Delayed'].mean()

df_clean['How_Long_Delayed'] = df_clean['How_Long_Delayed'].replace(0, mean_delay)


#df_clean['How_Long_Delayed']
df_clean['How_Long_Delayed']
# check the replacement
print('Number of empty entries in How Long Delayed:', df_clean.isnull().all(axis=1).sum())
print()
print()
print(df_clean['How_Long_Delayed'])
print()
print()
print('Print only raw where the replacement has been done', df_clean[df_clean['How_Long_Delayed'] == mean_delay]['How_Long_Delayed'])
```

```python
# Then convert back our how_long_delayed column to integer datatype
# ---
#
# first print Not a Number(NaN) value
#print('Print the raw with NAN', df_clean[df_clean['How_Long_Delayed'].isna()].sum())
# Fill NaN values with the mean value
df_clean['How_Long_Delayed'] = df_clean['How_Long_Delayed'].fillna(mean_delay)
df_clean['How_Long_Delayed']=df_clean['How_Long_Delayed'].astype(int)

print(df_clean.dtypes)

print(df_clean)

```
